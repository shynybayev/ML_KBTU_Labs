```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('talking-tom.png', -1)
plt.imshow(img)
```

```python
plt.imshow(img, interpolation='nearest')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img_resized = cv2.resize(img,(300,600))
plt.imshow(img_resized, 'gray')
```

```python
(h, w) = img.shape[:2]
# calculate the center of the image
center = (w / 2, h / 3)
tom_90 = 90
tom_180 = 180
tom_270 = 270
 
scale = 1.0

M = cv2.getRotationMatrix2D(center, tom_180, scale)
rotated270 = cv2.warpAffine(img, M, (w, h))

plt.imshow(rotated270, 'gray')
```

```python
from scipy import ndimage
kern=np.ones((3,3))
filtered = ndimage.convolve(img[:,:,0], kern, mode='nearest', cval=0.0)
plt.imshow(filtered,cmap='gray')
```
